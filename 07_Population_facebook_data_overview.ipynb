```python
more_men = 0
more_women = 0
same = 0
for index, row in data_esp.iterrows():
    #print(row["men"], row["women"], row["men"] > row["women"])
    if int(row["men"]) > int(row["women"]):
        more_men += 1
    elif row["women"] > row["men"]:
        more_women +=1
    elif row["women"] == row["men"]:
        same +=1

print()    
print("Barcelona")
print("Labs with more men:",more_men)
print("Labs with more women:",more_women)
print("Labs with same men and women:",same)
```
